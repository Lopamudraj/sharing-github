```python
# Simple Line Plot with plotly.express

import plotly.express as px

df = px.data.gapminder().query("country=='Canada'")
fig = px.line(df, x="year", y="lifeExp", title='Life expectancy in Canada')

fig.show()
```

```python
# Line Plot with column encoding color

import plotly.express as px

df = px.data.gapminder().query("continent !='Asia'")
fig = px.line(df, x="year", y="lifeExp", title='Life expectancy in Oceania',color = 'continent')

fig.show()
```

```python
#Line Plot with go.Scatter
    # Simple Line Plot

import plotly.graph_objects as go
import numpy as np


x = np.arange(10)
print(x)
fig = go.Figure(data=go.Scatter(x=x, y=x**7))

fig.show()  

```

```python
#Line Plot Modes

import plotly.graph_objects as go
import numpy as np

x = np.random.seed(1)

N = 100
random_x = np.linspace(0, 1, N)
random_y0 = np.random.randn(N) + 5
random_y1 = np.random.randn(N)
random_y2 = np.random.randn(N) - 5

fig = go.Figure()
fig.add_trace(go.Scatter(x=random_x, y=random_y0, mode='lines', name='lines'))
fig.add_trace(go.Scatter(x=random_x, y=random_y1, mode='lines+markers', name='lines+markers'))
fig.add_trace(go.Scatter(x=random_x, y=random_y2,line=dict(color='green',dash='dash')))
fig.add_trace(go.Scatter(x=random_x, y=random_y2, mode='markers', name='markers',line=dict(color='orange')))



fig.show()
```
